```python
import os
from crewai import Agent, Task, Crew, Process
from crewai_tools import SerperDevTool, tool
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
from google.colab import userdata
```

```python
os.environ['GOOGLE_API_KEY'] = userdata.get('GOOGLE_API_KEY')
os.environ['SERPER_API_KEY'] = userdata.get('SERPER_API_KEY')
```

```python
search_tool = SerperDevTool()
```

```python
llm= ChatGoogleGenerativeAI(
    model = "gemini-1.5-flash",
    verbose = True,
    temperature = 0.5,
    google_api_key = os.getenv('GOOGLE_API_KEY')
)

researcher = Agent(
    role = "Large Language Model Defense Research Analyst",
    goal = "Investigate and report on the latest innovations in Large Language Models in the defense industry in 2024.",
    backstory = """You work for a defense research institute.
    Your expertise lies in sourcing and analyzing information on Large Language Models that could be used in defense.""",
    verbose = False,
    allow_delegation = False,
    tools = [search_tool],
    llm = llm)

writer = Agent(
    role = "Defence Technology Writer and Coder",
    goal = "Craft informative articles on the latest innovations in Large Language Models in the defense industry in 2024.",
    backstory = """You are a technical writer writing informative articles tailored to developers.
    Your expertise lies in transforming complex Large Language Model concepts into a well-written background, code and commentary, and summary.""",
    verbose = True,
    allow_delegation = True,
    llm = llm)
```

```python

task1 = Task(
    description = "Investigate the most recent breakthroughs in Large Language Model applications for defense.",
    expected_output = "A detailed summary of the latest innovations in Large Language Model applications for defense sector.",
    agent = researcher
)

task2 = Task(
    description = "Compose an informative article highlighting the latest innovations in Large Language Model applications for defense and share example code for the reader to test the application.",
    expected_output = "An engaging and well-structured article on the recent advancements in Large Language Models for defense.",
    agent = writer
)
```

```python

crew = Crew(
    agents = [researcher, writer],
    tasks = [task1, task2],
    verbose = 1
)
```

```python
final = crew.kickoff()
```
